# Notebook 5: Running DA code to create a new Holocene reconstruction

In this activity, we'll be creating a new Holocene temperature reconstruction by running data assimilation code.

## 1. Setting things up

The data assimilation code and data can be downloaded here: https://drive.google.com/file/d/17ZJ83P7psrBCjieegiGdLML9H5Ol8A_L/view?usp=sharing.

After downloading that code, unzip it. Now, upload the new "data_assimilation" folder to your Google Drive, in the same folder as the other notebooks (the "da_activities" folder).

Next, open the file "data_assimilation/da_main_code.ipynb". You'll find the rest of the instructions there.

Once you've run the data assimilation, proceed to Notebook 6.
